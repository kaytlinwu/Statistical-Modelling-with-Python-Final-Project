Build a regression model.

```python
#import libraries required
import statsmodels.api as sm
import pandas as pd
import numpy as np
```

```python
#load df into notebook
df = pd.read_csv('../data/final_miami_bikes.csv')
```

```python
#Drop bike_station_coordinates column. 
#Will not be used in analysis, adds many duplicate rows that are unnecessary to building final model.
df_drop = df.drop(columns= ['Bike_Station_Coordinates'], axis=1)
```

```python
#Drop duplicate values
df_final = df_drop.drop_duplicates()
```

```python
#Drop all null values (cannot be used in the analysis)
df_final = df_final.dropna(how='any',axis=0) 
```

```python
#Verify that nf has no null values
df_final.isna().sum()
```

```python
df_final.shape
#df.shape
```

```python
df_final.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.pairplot(data=df_final)
plt.show()
```

From the pair plot above, does not look like there are any obvious correlations or linearity between the variables. 
A few of the graphs are seen to be slightly skewed. There are a few outliers as well. These outliers will not be removed in the analysis. 

```python
#Finding the correlation 
df_final.corr(numeric_only = True)
```

```python
#Check if the correlations are significant using Pearson's correlation coefficient test

import scipy

stat, p1 = scipy.stats.pearsonr(df_final['Popularity'], df_final['bikes'])
stat, p2 = scipy.stats.pearsonr(df_final['Price'], df_final['bikes'])
stat, p3 = scipy.stats.pearsonr(df_final['Rating'], df_final['bikes'])

print(p1, p2, p3)
```

As seen above, the correlations are not very strong. The strongest correlation can be seen between the 'Rating' and the 'bikes column with a p value of ~0.50

```python
#Performing model using 1 variable to start
#choosing df['Rating'] as the x variable as it seems to have the highest correlation 

y = df_final['bikes']
x = df_final['Rating']
#x = sm.add_constant(x) 
#constant was removed because it reduced the R_squared value by a significant amount.
x.head()  #view both x and y
y.head()
```

```python
#Create the model using x and y 

model = sm.OLS(y,x)
results = model.fit()  #fit the model 
```

Provide model output and an interpretation of the results. 

```python
print(results.summary())
```

The output using 1 independent variable, rating, was seen to have a strong R-squared of 0.595 meaning that 59.5% of the data could be correlated to the regression that was made by the model. 
Additionally, the p value was seen to be significant with a p value of 0.000. As well there was a coefficient varaible that came up as 0.6966. 

```python
#To test for normality
#create residual plots
residuals = results.resid
residuals
```

The majority of the residual points are seen to have a negative correlation to the regression. 
Let us visualize the residuals in a histogram as seen below. 

```python
import seaborn as sns
import matplotlib.pyplot as plt

#visualize residuals as a histogram
sns.histplot(residuals)
```

The histogram looks like the values are heavily skewed to the left meaning that a majority of data seem to be negatively correlated to the linear regression
A few points are seen to be heavily to the right. 

```python
#visualize the data as a scatterplot
sns.scatterplot(data=df_final, x= 'Rating', y='bikes', hue= 'Place_Type')
```

```python
#To see if bikes and rating come from a population with the same distribution.
sm.qqplot(residuals, line='r')
plt.show()
```

Looking at the qq plot, the points are somewhat aligned to a few areas on the reference line. The 2 tail ends of the plot do not seem to match.
This suggests that the populations may belong to the same distribution with a low degree of strength.

```python
scipy.stats.shapiro(residuals)
```

As the p value is below 0.05, this means that the Shapiro test highly suggest that the data is notnormally distributed

```python
#homoscedasticity tests visual representation using scatterplot

ax= sns.scatterplot(x=df_final['Rating'], y= residuals)
ax.axhline(0)  #create a horizontal line at 0
```

Adding more independent variables to the model to make a multiregression model. 

```python
#Set x and y variables for model
#y = df_dummify['bikes']
y = np.log(df_final['bikes']+1)   #normalized data by taking log of y
x = df_final[['Popularity',
                'Rating']]
#x = sm.add_constant(x) #removed constant as it decreased the R_squared value
x.head()  
```

```python
model = sm.OLS(y,x)
results = model.fit()  #fit the model 
print(results.summary())
```

Compared to the previous model summary, the multiregression model have a stronger R-squared value. 
Interestingly, the p value for each independent variable are above 0.05, with the exception of Locality_Miami, meaning that the independ variables by themself are not statistically significant. The coefficient for each variable seen to be quite large ranging from -2.855 to 3.3363. 

```python
#To test for normality of the multiregression model
#create residual plots
residuals = results.resid
residuals
```

Residuals are negative but are less than previously calculated. 

```python
#visualize residuals as a histogram
sns.histplot(residuals)
```

The residuals are seen to be more normally distributed compared to using 1 variable to create the model

```python
#To see if variables are within the same population 
sm.qqplot(residuals, line='r')
plt.show()
```

qq plot is very tight to the line with slight deviations at the ends of the graph. 

```python
#Formal test for homoscediscity 
scipy.stats.shapiro(residuals)
```


Shapiro p value is significantly below 0.05. Suggests that the data is not drawn from a normal distribution. 

```python
#bresuch test for homoscdiscity

stat, p, f_stat, f_p = sm.stats.diagnostic.het_breuschpagan(residuals, results.model.exog)
print(p, f_p)
```

From the bresuchpagan test we can say that the model is not normally distributed. 

```python
#Create the model predictor here:
m_lst = results.params[['Popularity',
                'Rating']]
print(m_lst)
```
